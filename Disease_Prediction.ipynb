# Training by Naive Bayes

```python
class MultinominalNB():

  def __init__(self, alpha = 1):
    self.alpha = alpha

  def prob_class(self, c):
    count_c = sum([self.y_train[i] == c for i in range(self.total)])
    return count_c/self.total

  def prob_feature_class(self, f, c):
    count_f = 0
    count_c = 0
    for i in range(self.total):
      if self.y_train[i] == c:
        if self.X_train[i, f] == 1:
          count_f += 1
        count_c += 1
    if count_f == 0:
      count_f = 1
      count_c += len(self.features)
    return count_f/count_c
  
  def fit(self, X_train, y_train, feature, prognosis):
    self.total = len(X_train)
    self.X_train = X_train
    self.y_train = y_train
    self.features = feature
    self.classes = prognosis
    n_classes = len(self.classes)
    n_features = len(self.features)
    self.arr_prob_class = np.zeros(n_classes)
    self.arr_prob_feature_class = np.zeros((n_classes, n_features))
    for i in range(n_classes):
      self.arr_prob_class[i] = self.prob_class(self.classes[i])
      for j in range(n_features):
        self.arr_prob_feature_class[i, j] = self.prob_feature_class(j, self.classes[i])
    self.all_symptoms()

  def prob_test_class(self, test, c):
    prob_test_log2 = log2(self.arr_prob_class[self.classes.index(c)])
    for i in range(len(self.features)):
      prob_feature = self.arr_prob_feature_class[self.classes.index(c), i]
      if test[i] == 1:
        prob_test_log2 += log2(prob_feature)
      else:
        if prob_feature < 1:
          prob_test_log2 += log2(1 - prob_feature)
        else:
          prob_test_log2 += log2(1/(sum([self.y_train[i] == c for i in range(self.total)]) + len(self.features)))

    return prob_test_log2

  def predict_test(self, test):
    list_prob = [self.prob_test_class(test, c) for c in self.classes]
    return self.classes[list_prob.index(max(list_prob))]
  
  def predict_test_near(self, test):
    list_prob = [self.prob_test_class(test, c) for c in self.classes]
    first_position = list_prob.index(max(list_prob))
    list_prob.pop(first_position)
    second_result = self.classes[list_prob.index(max(list_prob))]
    second_position = list_prob.index(max(list_prob))
    list_prob.pop(second_position)
    third_result = self.classes[list_prob.index(max(list_prob))]
    return second_result, third_result

  def accuracy(self, X_test, y_test):
    arr_predict_test = np.array([self.predict_test(X_test[i]) for i in range(len(X_test))])
    return np.sum(y_test == arr_predict_test)/len(y_test)

  def symptoms(self, disease):
    first_flag = True
    symptoms = np.zeros(len(self.features))
    for i in range(self.total):
      if y_train[i] == disease:
        if first_flag:
          symptoms = X_train[i]
          first_flag = False
        else:
          symptoms += X_train[i]
    return list(self.features[i] for i in range(len(self.features)) if symptoms[i] > 0)

  def all_symptoms(self):
    self.list_all_symptom = []
    for disease in self.classes:
      self.list_all_symptom.append(self.symptoms(disease))

  def get_symptoms(self, disease):
    return self.list_all_symptom[self.classes.index(disease)]
```

```python
model = MultinominalNB()
model.fit(X_train, y_train, list_feature, list_prognosis)
```

# Testing

```python
X_test = data_test.drop(['prognosis'], axis=1)
y_test = data_test['prognosis']
X_test = np.array(X_test)
y_test = np.array(y_test)
model.accuracy(X_test, y_test)
```

# Data processing for testing input

```python
# Predict disease which causes list_symptom
def predict_input_symptoms(list_symptom):
  # Convert to np array
  np_symptom = np.zeros(len(list_feature))
  for symptom in list_symptom:
    if symptom in list_feature:
      np_symptom[list_feature.index(symptom)] = 1
  # Predict disease
  disease = model.predict_test(np_symptom)
  return disease
```

```python
# Get disease near predicted disease which causes list_symptom
def predict_near_input_symptoms(list_symptom):
  # Convert to np array
  np_symptom = np.zeros(len(list_feature))
  for symptom in list_symptom:
    if symptom in list_feature:
      np_symptom[list_feature.index(symptom)] = 1
  # Predict disease
  second_disease, third_disease = model.predict_test_near(np_symptom)
  return second_disease, third_disease
```

# Predicting from chatbox

```python
# Get symptoms of a disease
def get_symptoms_disease(disease):
  return model.get_symptoms(disease)
```

```python
# Get difference symptoms
def difference_symptoms(list_symptom_1, list_symptom_2):
  difference = set(list_symptom_2).difference(list_symptom_1)
  return list(difference)
```

```python
# Compare result
def compare_result(result_1, result_2):
  if result_1 != result_2:
    return False
  return True
```

```python
# List symptom from chatbox
# list_symptom = ['itching',	'skin_rash',	'nodal_skin_eruptions']

output_disease = "Can't predict!"
previous_result = output_disease
num_predict = 0

while num_predict < 5:
  # Prediction
  
  # Result
  result = predict_input_symptoms(list_symptom)
  print(result)

  # Compare result
  if compare_result(previous_result, result):
    output_disease = result
    break
  # Get symptoms of predicted disease
  list_symptom_result = get_symptoms_disease(result)

  # Get symptoms to ask more
  ask_more = difference_symptoms(list_symptom, list_symptom_result)
  if len(ask_more) == 0:
    output_disease = result
    break
  print(ask_more)

  # List new symptom --- GET FROM CHATBOX
  string_next = input("More symptoms: ")
  list_new_symptom = []
  viet_to_eng(list_new_symptom, string_next)
  if len(list_new_symptom) == 0:
    second_result, third_result = predict_near_input_symptoms(list_symptom)
    list_symptom_second_result = get_symptoms_disease(second_result)
    ask_more = difference_symptoms(list_symptom, list_symptom_second_result)
    print(second_result)
    print(ask_more)
    string_next = input("More symptoms: ")
    viet_to_eng(list_new_symptom, string_next)
    if len(list_new_symptom) == 0:
      list_symptom_third_result = get_symptoms_disease(third_result)
      ask_more = difference_symptoms(list_symptom, list_symptom_third_result)
      print(third_result)
      print(ask_more)
      string_next = input("More symptoms: ")
      viet_to_eng(list_new_symptom, string_next)
    
  # List symptom after updating
  list_symptom += list_new_symptom
  previous_result = result
  num_predict += 1

if output_disease == "Can't predict!":
  second_result, third_result = predict_near_input_symptoms(list_symptom)
  print("Chúng tôi không thể chẩn đoán chính xác")
  print("Bạn có khả năng mắc một trong ba bệnh sau:", previous_result, second_result, third_result)
else:
  print("Bạn bị mắc bệnh:", output_disease)
```
